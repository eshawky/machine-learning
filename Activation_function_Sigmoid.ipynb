```python
def my_sigmoid(x):
  f=1/(1+np.exp(-x))
  return f
```

```python
x=np.arange(-10,10,0.1)
y=my_sigmoid(x)
y_div=y*(1-y)
X_norm=x/10
plt.plot(x,y)
plt.plot(x,y_div)
#lt.plot(x,X_norm)
plt.grid()
```
